```python
all = data_all.copy()
all_sm = all['SMILES']

anti = data_anti.copy()
anti_sm = anti['SMILES']
```

```python
commmon_count = anti_sm.isin(all_sm).sum()
print(commmon_count)
```

```python
smiles_excluded = all_sm[~all_sm.isin(anti_sm)]
```

```python
smiles_excluded = pd.DataFrame(smiles_excluded)
anti_sm = pd.DataFrame(anti_sm)

smiles_excluded['anti_cancer'] = 0
anti_sm['anti_cancer'] = 1
```

```python
df = pd.concat([smiles_excluded, anti_sm], ignore_index=True)
df
```
